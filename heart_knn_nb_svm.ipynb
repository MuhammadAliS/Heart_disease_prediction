```python
#Spliting the dataset for training data and testing data
from sklearn.model_selection import train_test_split
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=.2)
```

```python
#Scaling the features for greater accuracy
from sklearn.preprocessing import MinMaxScaler
scalar= MinMaxScaler()
scalar.fit(x_train)
scalar.transform(x_train)
scalar.transform(x_test)
print('Done')
```

```python
#Importing required modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# FINDING BEST VALUE FOR K 
ac=0
bc=100
for i in range(1,50):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    a=accuracy_score(y_test,y_pred)
    b=round(a*100)
    if(b>ac):
        ac=b
        n=i
    
print(f'Maximum accuracy:   {ac}%')
print(f'K value         :   {n}')
```

```python
#Training and evaluating the Model
def hpt(x_train,x_test,y_train,y_test):
    model1=KNeighborsClassifier(n_neighbors=1)
    model2=GaussianNB()
    model3=SVC()
    model1.fit(x_train,y_train)
    model2.fit(x_train,y_train)
    model3.fit(x_train,y_train)
    y_pred1=model1.predict(x_test)
    y_pred2=model2.predict(x_test)
    y_pred3=model3.predict(x_test)
    a1,a2,a3=accuracy_score(y_test,y_pred1),accuracy_score(y_test,y_pred2),accuracy_score(y_test,y_pred3)
    print(f'KNN accuracy        :  {round(a1*100)}%')
    print(f'Naive Bayes accuracy:  {round(a2*100)}%')
    print(f'SVM accuracy        :  {round(a3*100)}%')
    return y_pred1,y_pred2,y_pred3
print('Accuracy Score')
print('...............\n')
y1,y2,y3=hpt(x_train,x_test,y_train,y_test)
print('\nClassification Report of KNN')
print('...............\n')
c=classification_report(y_test,y1)
print(c)
print('\nClassification Report of Naive Bayes')
print('...............\n')
c=classification_report(y_test,y2)
print(c)
print('\nClassification Report of SVM')
print('...............\n')
c=classification_report(y_test,y3)
print(c)
```
